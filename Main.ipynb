```python
import pandas as pd
```

```python
# Load data
calendar_df = pd.read_excel("AdventureWorks_Database.xlsx", sheet_name="Calender")
customer_df = pd.read_excel("AdventureWorks_Database.xlsx", sheet_name="Customers")
product_df = pd.read_excel("AdventureWorks_Database.xlsx", sheet_name="Product")
territory_df = pd.read_excel("AdventureWorks_Database.xlsx", sheet_name="Territory")
sales_df = pd.read_excel("AdventureWorks_Database.xlsx", sheet_name="Sales")
budget_df = pd.read_excel("Budget.xlsx", skiprows=3)
```

```python
# Explore data
print("Calendar Data Info:")
print(calendar_df.info())
print("\nCustomer Data Info:")
print(customer_df.info())
print("\nProduct Data Info:")
print(product_df.info())
print("\nTerritory Data Info:")
print(territory_df.info())
print("\nSales Data Info:")
print(sales_df.info())
print("\nBudget Data Info:")
print(budget_df.info())

# Check for missing values
print("\nMissing Values in Calendar Data:")
print(calendar_df.isnull().sum())
print("\nMissing Values in Customer Data:")
print(customer_df.isnull().sum())
print("\nMissing Values in Product Data:")
print(product_df.isnull().sum())
print("\nMissing Values in Territory Data:")
print(territory_df.isnull().sum())
print("\nMissing Values in Sales Data:")
print(sales_df.isnull().sum())
print("\nMissing Values in Budget Data:")
print(budget_df.isnull().sum())

# Handling missing values
# Example: Filling missing values with mean
calendar_df.fillna(calendar_df.mean(), inplace=True)
customer_df.fillna(customer_df.mean(), inplace=True)
product_df.fillna(product_df.mean(), inplace=True)
territory_df.fillna(territory_df.mean(), inplace=True)
sales_df.fillna(sales_df.mean(), inplace=True)
budget_df.fillna(budget_df.mean(), inplace=True)

# Check for duplicates
print("\nDuplicate Rows in Calendar Data:")
print(calendar_df.duplicated().sum())
print("\nDuplicate Rows in Customer Data:")
print(customer_df.duplicated().sum())
print("\nDuplicate Rows in Product Data:")
print(product_df.duplicated().sum())
print("\nDuplicate Rows in Territory Data:")
print(territory_df.duplicated().sum())
print("\nDuplicate Rows in Sales Data:")
print(sales_df.duplicated().sum())
print("\nDuplicate Rows in Budget Data:")
print(budget_df.duplicated().sum())
```

```python
# Calculate key metrics
total_sales = sales_df['SalesAmount'].sum()
average_sales = sales_df['SalesAmount'].mean()
top_selling_products = sales_df.groupby('ProductKey')['SalesAmount'].sum().sort_values(ascending=False).head(5)

# Customer demographics
average_income_by_education = customer_df.groupby('Education')['YearlyIncome'].mean()
average_income_by_occupation = customer_df.groupby('Occupation')['YearlyIncome'].mean()

# Analyze factors affecting sales and budget variances 
budget_total = budget_df[['Jan, 2016', 'Feb, 2016', 'Mar, 2016', 'Apr, 2016', 'May, 2016', 'Jun, 2016', 'Jul, 2016', 'Aug, 2016', 'Sep, 2016', 'Oct, 2016', 'Nov, 2016', 'Dec, 2016']].sum().sum()
variance = total_sales - budget_total

# Display key metrics and factors
print("Key Metrics:")
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {average_sales}")
print("\nTop Selling Products:")
print(top_selling_products)

print("\nCustomer Demographics:")
print("Average Income by Education:")
print(average_income_by_education)
print("\nAverage Income by Occupation:")
print(average_income_by_occupation)

print("\nFactors Affecting Sales and Budget Variances:")
print("Total Budget:")
print(budget_total)
print("Variance:")
print(variance)
```
